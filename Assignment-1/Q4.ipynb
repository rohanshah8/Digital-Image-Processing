```python
def connected_component(img):
  th=threshold_otsu(img)
  binary=img>th
  binary=np.where(binary==True,255,0)
  p=0
  q=deque()
  height=img.shape[0]
  width=img.shape[1]
  visited=np.zeros(height*width).reshape(height,width)
  comp=np.zeros(height*width).reshape(height,width)
  total_pixels=[]
  for i in range(0,height):
    for j in range(0,width):

      if (not visited[i][j]) and binary[i][j]==255:
        p+=1
        s=0
        q.append((i,j))
        while len(q)>0:
          s+=1
          x,y=q.popleft()
          visited[x][y]=True
          comp[x][y]=p

          if (x-1>=0 and binary[x-1][y]==binary[x][y] and (not visited[x-1][y])):
            visited[x-1][y]=True
            q.append((x-1,y))
          if (x+1<height and binary[x+1][y]==binary[x][y] and (not visited[x+1][y])):
            visited[x+1][y]=True
            q.append((x+1,y))
          if (y-1>=0 and binary[x][y-1]==binary[x][y] and (not visited[x][y-1])):
            visited[x][y-1]=True
            q.append((x,y-1))
          if (y+1<width and binary[x][y+1]==binary[x][y] and  (not visited[x][y+1])):
            visited[x][y+1]=True
            q.append((x,y+1))
        total_pixels.append(s)
  pi=total_pixels[1] #no. of pixels in digit 1
  pixels_digit=Counter(total_pixels)
  return len(total_pixels),pixels_digit[pi]


  
```
