```python
from cv2 import cv2
import matplotlib.pyplot as plt
import numpy as np
import skimage.io as io
from skimage.filters import threshold_otsu
```

```python
def Foreground_Extraction(text,background):
  pi=np.divide(text,255)
  ni=np.subtract(1,pi)
  background[:,:,2]=np.multiply(background[:,:,2],ni)
  background[:,:,1]=np.multiply(background[:,:,1],ni)
  background[:,:,0]=np.multiply(background[:,:,0],ni)
  background[:,:,0]=np.add(np.multiply(pi,255),background[:,:,0])
  return background
```

```python
path='/content/drive/MyDrive/DIP/Assignment01/'
gray_img=io.imread(path+'input/SingleColorText_Gray.png',as_gray=True)
img2=io.imread(path+'input/GrassBackground.png')
th=threshold_otsu(gray_img)
text=gray_img
np.where(text>=th,255,0)
newimg=Foreground_Extraction(text,img2)
plt.imshow(newimg)
plt.imsave(path+'output/text_background.png',newimg)
```
