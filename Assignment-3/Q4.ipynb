```python
import numpy as np
from scipy.optimize import minimize_scalar 
```

```python
g=np.zeros((5,5))
g[0,2]=1
g[1,1]=1
g[1,2]=2
g[1,3]=1
g[2,0]=1
g[2,1]=2
g[2,2]=-16
g[2,3]=2
g[2,4]=1
g[3,1]=1
g[3,2]=2
g[3,3]=1
g[4,2]=1

H=np.zeros((5,5))
for i in range(0,5):
  for j in range(0,5):
    H[i,j]=(i-2)**2+(j-2)**2
```

```python
def optimal_k1(k):
  h=np.fft.ifft2(np.fft.ifftshift(k*H))
  
  return abs(np.sum(np.square(np.subtract(h,g))))
```

```python
out1=minimize_scalar(optimal_k1)
```
